```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import time
```

```python
# Define the Arenstorf problem
def arenstorf(t, y):
    mu = 0.012277471

    y1, y2, y3, y4 = y

    r1 = np.sqrt((y1 + mu)**2 + y2**2)
    r2 = np.sqrt((y1 - 1 + mu)**2 + y2**2)

    dy1 = y3
    dy2 = y4
    dy3 = y1 + 2 * y4 - ((1 - mu) * (y1 + mu)) / r1**3 - (mu * (y1 - 1 + mu)) / r2**3
    dy4 = y2 - 2 * y3 - ((1 - mu) * y2) / r1**3 - (mu * y2) / r2**3

    return [dy1, dy2, dy3, dy4]
```

```python
# Set initial conditions
initial_conditions = [0.994, 0, 0, -2.00158510637908252240537862224]

# Set up time span
Tmax_periodic = 17.0652165601579625588917206249
Tmax_long = 100
```

## RK45 method

```python
# Solve using RK45 for the periodic orbit
start_time = time.time()
sol_periodic = solve_ivp(arenstorf, [0, Tmax_periodic], initial_conditions, method='RK45', rtol=1e-12, atol=1e-12)
end_time = time.time()

print(f"CPU time for DOPRI5(4): {end_time - start_time:.4f} seconds")

# Plot the periodic orbit
plt.figure(figsize=(8, 6))
plt.plot(sol_periodic.y[0], sol_periodic.y[1], label='RK45')
plt.title('Arenstorf Problem - Periodic Orbit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Solve using RK45 for long integration
start_time = time.time()
sol_rk45 = solve_ivp(arenstorf, [0, Tmax_long], initial_conditions, method='RK45', rtol=1e-12, atol=1e-12)
end_time = time.time()

print(f"CPU time for RK45: {end_time - start_time:.4f} seconds")

# Plot the orbit for RK45
plt.figure(figsize=(8, 6))
plt.plot(sol_rk45.y[0], sol_rk45.y[1], label='RK45')
plt.title('Arenstorf Problem - RK45')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
